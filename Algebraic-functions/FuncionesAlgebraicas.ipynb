# ***Algebraic Functions with Python***

```python
## Variables del codigo
import matplotlib.pyplot as plt
import numpy as np
```

## ***Types of Functions***

### **Linear function**
- The linear functions have the form: $f(x)=mx+b$.
- Wich: $m | b ∈ R$.
- The slope of the linear function is calculed as: $m=\frac{y_2-y_1}{x_2-x_1}$ or substracting the positions of X and Y, and the point of cut on the axis Y represented by 'b'.
- Its domain is: $Dom_f=-∞,∞$ and its image is: $Im_g=-∞,∞$.

```python
## Algebraic linear function
resolution = 100
def f(x, m, b):
  return m * x + b

x = np.linspace(-10, 10, resolution)
y = f(x, 10, 5) #Vars: x -> Slope, b -> Scatter point in the Y axis

#>> Chart
plt.plot(x, y, label="$f(x) = mx + b$")
plt.title('Linear function')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(loc='upper left')
plt.grid()

plt.show()
```

### **Polynomial function**
- Polyninomial functions have the form: $Pol(x)= a_n x^n+ a_{n-1} x^{n-1}= a_2 x^2+ a_1 x$.
- A function with this form is called polynome of $n$ grade (The biggest number).
- We'll call the elements 'Coefficients' which: $a ∈R$.

- For example: $P(x)=2x^7-x^4+3x^2+4$, it's a polynome of 7th grade, this is the biggest index in the term.

```python
## Polynomial Function
def f(x):
  return (2*x**7)-(x**4)+(3*x**2)+4
y = f(x)

#>> Chart
plt.plot(x, y, label="$P(x)=2x^7-x^4+3x^2+4$")
plt.title('Polynomial function')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(loc='upper left')
plt.grid()

plt.show()
```

#### ***Note***
When we use the terms: $P(x)=2x^7-x^4+3x^2+4$ and we change the first exponent to:
- An even number, the form of the function it takes an $U$ form.
- An odd number, the function's form will be like an ascendent wave.

### ***Power functions***
- Similar to Polynomial functions, they have the form:$f(x) = x^2$.
- Its domain is: $Dom_f = (-∞, ∞)$.
- Its image is: $Im_f = (0, ∞)$

- For example: $f(x) = x^2$.

```python
## Power function
def f(x):
  return x**7
y = f(x)

#>> Chart
plt.plot(x, y, label="$P(x)=x^7$")
plt.title('Power function')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(loc='upper left')
plt.grid()

plt.show()
```

### ***Transcendent algebraic functions***

#### ***Trigonometrical functions***
Some examples are: $sin(x), cos(x)$ y $tan(x)$.

```python
## Trigonometrical functions
def f(x):
    return np.sin(x), np.cos(x), np.tan(x)
y_sin, y_cos, y_tan = f(x)

#>> Canvas
fig, ax = plt.subplots(3, 1, figsize=(8, 8), label=('Funciones trigonometricas'))

#> Sin(x)
ax[0].plot(x, y_sin, label="$Sin(x)$")
ax[0].set_title('Sine of X')
ax[0].set_xlabel('X')
ax[0].set_ylabel('Y')
ax[0].legend(loc='upper left')
ax[0].grid()

#> Cos(x)
ax[1].plot(x, y_cos, label="$Cos(x)$")
ax[1].set_title('Cosine of X')
ax[1].set_xlabel('X')
ax[1].set_ylabel('Y')
ax[1].legend(loc='upper left')
ax[1].grid()

#> Sin(x)
ax[2].plot(x, y_tan, label="$Tan(x)$")
ax[2].set_title('Tangent of X')
ax[2].set_xlabel('X')
ax[2].set_ylabel('Y')
ax[2].legend(loc='upper left')
ax[2].grid()

fig.tight_layout() #Espaciado entre graficas
plt.show() #Mostrar grafica

```

#### ***Exponential functions***
- The have the form: $f(x) = a^x$.
- Which the base $a$ is a positive constant. For example, we can use the base like the $euler's$ number.

$f(x) = e^2$

```python
## Exponential functions
def f(x):
    return np.exp(x)
y = f(x)

#>> Charts
plt.plot(x, y, label='$f(x) = e^2$')
plt.title('Exponential function')
plt.ylabel('Y')
plt.xlabel('X')
plt.legend()
plt.grid()

plt.show()
```

#### ***Logarithmic functions***
The logarithm is defined by the **relation**:  $log_b(x) = n$ **->** $x = b^n$

Which:
- $b$ **->** Is the base.
- $n$ **->** Is the exponent that is powering the base.
- $x$ **->** Is the result of powering the base with the exponent: $b^n$.

***Example:*** We have: $b=2$ y $n=8$, so:  $2^8 = 256$.
- So: $x=256$. Calculating the logarithm with base 2 of $x$ is: $log_2(256) = 8$.

```python
## Logarithmic functions
def f(x):
    return np.log2(x)

x = np.linspace(0.001, 256, 1000) #The first number has to be different from 0
y = f(x)

#>> Chart
plt.plot(x, y, label='$log_2(256) = 8$')
plt.title('Logarithmic function')
plt.ylabel('Y')
plt.xlabel('X')
plt.legend()
plt.grid()

plt.show()
```

### ***Seccioned functions***
Ther're functions that have different defined values from an interval. For example, the step function of Heaviside:

$H(x) = {{0, x < 0 | 1, x > 0}}$

```python
## Seccioned functions
def H(x):
    v = np.zeros(len(x))
    for idx, x in enumerate(x):
        if x > 0:
            v[idx] = 1
        else:
            v[idx] = 0
    return v        
    
n = 1000
x = np.linspace(-10, 10, n)    
        
y = H(x)

#>> Charts
plt.plot(x, y, label='$H(x) = 0, x<0|1, x>0$')
plt.title('Seccioned functions')
plt.ylabel('Y')
plt.xlabel('X')
plt.legend()
plt.grid()

plt.show()
```

### ***Compound functions***
The compound functions are 2 or more algebraic functions:
$f-o-g = f(g(x))$

```python
## Compound functions
n = 1000
x = np.linspace(-10, 10, n)

def g(x):
    return x**2

def f(x):
    return np.sin(x)

f_o_g = f(g(x)) #Combine the functions from variables

##> Chart
plt.plot(x, f_o_g, label='f(g(x))')
plt.title('Compound functions')
plt.ylabel('Y')
plt.xlabel('X')
plt.legend()
plt.grid()

plt.show()
```

## ***Function manipulation***

### ***Horizontal and vertical Movements***


Being $c$ a constant bigger than 0, then:
- $y = f(x) + c$, it moves $c$ unities up.
- $y = f(x) - c$, it moves $c$ unities down.
- $y = f(x + c)$, it moves $c$ unities right.
- $y = f(x - c)$, it moves $c$ unities left.

```python
## Horizontal and vertical Movements
n = 1000
x = np.linspace(-10, 10, n)
c = 5 #Numero de unidades que nos queremos desplazar

def f(x):
    return x**2
y = f(x) - c

##> Chart
fig, ax = plt.subplots()
ax.plot(x, y,label='$f(x) = m * x + b$')
ax.set_title('Horizontal and vertical Movements')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.grid()
ax.legend()

ax.axvline(x=0, color='g')
ax.axhline(y=0, color='g')

plt.show()
```

### ***Enlargements and compressions***

Being $c$ a constant bigger than 0, then:
- $y = c * f(x)$, enlarges the function vertically in a factor of $c$.
- $y = \frac{1}{c} * f(x)$, compress the function vertically in a factor of $c$.
- $y = f(c * x)$, enlarges the function horizontally in a factor of $c$.
- $y = f(\frac{1}{c} * x)$, compress the function horizontally in a factor of $c$.

```python
## Enlargements and compressions
n = 1000
def f(x):
    return np.sin(x)

c = 2
x = np.linspace(-15, 15, n)
y = f(1/c * x)

#>> Chart
fig, ax = plt.subplots()
ax.plot(x, y, label='$sin(x)$')
ax.set_title("Enlargements and compressions")
ax.grid()
ax.legend(loc='upper left')
ax.axhline(y=0, color='g')
ax.axvline(x=0, color='g')

plt.show()
```

### ***Reflection of algebraic functions***

Reflect a function:
- $y = -f(x)$, reflect the function with respect to $X$ axis.
- $y = f(-x)$, reflect the function with respect to $Y$ axis.

```python
## Reflection of algebraic functions
n = 1000
def f(x):
    return x**3

x = np.linspace(-100, 100, n)
y = -f(x) 

#>> Grafica de las funciones
fig, ax = plt.subplots()
ax.set_title('Reflection of algebraic functions')
ax.plot(x, y, label='$-f(x^3)$')
ax.grid()
ax.legend()
ax.axhline(y=0, color='g')
ax.axvline(x=0, color='g')

plt.show()
```
