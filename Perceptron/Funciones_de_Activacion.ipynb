# ***Activation functions with Perceptrons***

### *Linear funtcion*

```python
def f(x):
    return x

#>> Chart
plt.plot(x, f(x))
plt.grid()
plt.show()
```

### *Step function or Heaviside's*
When we use Perceptrons this function helps to clasify categories.

```python
def h(x):
    y = np.zeros(len(x))
    for idx, x in enumerate(x):
        if x>=0:
            y[idx]=1
    return y

y = h(x)

#>> Chart
plt.plot(x, y)
plt.grid()
plt.show()
```

### *Sigmoid function*
Serves to make a process called **Logistic Regression** and helps to talk about probabilities in a range. When the Sigmoid function is 0 the line'll be placed in **0.5**.

```python
def f(x):
    return 1/(1 + np.exp(-x))
y = f(x)

#>> Chart
plt.plot(x, y)
plt.grid()
plt.show()
```

### *ReLU function*
All the values smaller or equal to $0$ or a number $n$ will be represented with a 0, and the other numbers'll be represented with the value of $x$.

```python
def f(x):
    return np.maximum(x, 0)
y = f(x)

#>> Chart
plt.plot(x, y)
plt.grid()
plt.show()
```

# ***How to calculate a Quadratic Median Error function***
This function serve to calculate the error that can have our funcion when realized:

![Function-errors](https://raw.githubusercontent.com/XJDevX/myPythonNotes/main/Perceptron/imgs/error-calculate.png)
- In this case the line of the Model is called: $\hat{y}$.

This is because not all values will be in a straigth line. Thsi functions can be written as:
- Substracting the real value for the define function: $error = \hat{y} - y$, this way can generate errors when using negative values.
- Solving the previous problem: $|error| = |\hat{y} - y|$, a good way to do it but not the best.
- Other way to solve the problem is: $error^2 = (\hat{y} - y)^2$, the small errors won't be so visible.
- Obtain the total error of the function: $E = \sum_{i=1}^{n} (\hat{y} = y)^2$, isn't so useful as the individual error.
- Obtain the **Quadratic Median Error**: $QME = \frac{1}{n} \sum_{i=1}^{n} (\hat{y} - y)^2$, the best way to obtain the individual error.
