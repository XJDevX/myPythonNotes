### *Funcion lineal*

### *Funcion escalon o de Heaviside*
Al momento de utilizar perceptrones esta funcion nos puede ayudara hacer clasificaciones categoricas.

```python
# Generar la funcion
def h(x):
    y = np.zeros(len(x))
    for idx, x in enumerate(x):
        if x>=0:
            y[idx]=1
    return y

y = h(x)

#>> Grafica de la funcion
plt.plot(x, y)
plt.grid()
plt.show()
```

### *Funcion sigmoide*
Nos sirve para un proceso llamado **Regresion Logistica** y porque sirve para hablar de probabilidades en un rango. Cuando la funcion Sigmoide es 0 la linea estara ubicada en **0.5**.

```python
## Funcion
def f(x):
    return 1/(1 + np.exp(-x))

y = f(x)
#>> Grafica de la funcion
plt.plot(x, y)
plt.title('Funcion Sigmoide')
plt.grid()

plt.show()
```

### *Funcion ReLU*
Todos los valores menores o iguales a $0$ o un numero $n$ seran representados como un 0 y los numeros mayores a $0$ o a el numero $n$ seran representados con el valor de $x$.

```python
## Funcion
def f(x):
    return np.maximum(x, 0)

y = f(x)
#>> Grafica de la funcion
plt.plot(x, y)
plt.title('Funcion ReLU')
plt.grid()

plt.show()
```

# ***Como calcular una Funcion de Error Cuadratico Medio***
Esta funcion sirve para calcular el fallo que puede tener nuestra funcion al ser realizada: 

![Errores-Funciones](https://static.platzi.com/media/user_upload/ecm-d9d507ff-d053-4a9b-853d-11c9d6a3a261.jpg)

> En este caso la linea de Modelo es escrita como: $\hat{y}$.

Esto se debe a que no todos los valores van a ir en linea recta. Esta funcion se pueden describirse como:

> Restanto el valor real por la funcion definida: $error = \hat{y} - y$, esta manera puede generar error al manejar datos negativos.

> Solucionando la problematica anterior: $|error| = |\hat{y} - y|$, una manera buena pero no la mejor.

> Otra manera de solucionar el problema es: $error^2 = (\hat{y} - y)^2$, los errores pequeños no seran tan visibles.

> Conseguir el error total de la funcion: $E = \sum_{i=1}^{n} (\hat{y} = y)^2$, no es tan util como el error individual.

> Conseguir el **Error Cuadratico Medio**: $ECM = \frac{1}{n} \sum_{i=1}^{n} (\hat{y} - y)^2$, la mejor manera de conseguir el error individual.
