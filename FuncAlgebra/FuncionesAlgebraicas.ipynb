# ***Funciones algebraicas con Python***

```python
## Variables del codigo
import matplotlib.pyplot as plt
import numpy as np
```

## ***Tipos de funciones con Python***

### **Funcion lineal**
Las funciones lineales tienen la forma: $f(x)=mx+b$:

En el cual: $m | b ∈R$

La pendiente de la función lineal es calculada como: $m=\frac{y_2-y_1}{x_2-x_1}$ o restando las dos posiciones de X y Y, y el punto de corte con el eje Y es representado por b.

Su dominio es $Dom_f= -∞,∞$ y su imagen es $Im_f= -∞,∞$.


```python
## Funcion algebraica lineal
resolution = 100
def f(x, m, b):
  return m * x + b

x = np.linspace(-10, 10, resolution)
y = f(x, 10, 5) #Vars: x, Pendiente, Corte con el eje Y

#>> Grafica de la funcion
plt.plot(x, y, label="$f(x) = mx + b$")
plt.title('Funcion lineal')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(loc='upper left')
plt.grid()

plt.show() #Mostrar grafica
```

### **Funcion polinomica**
Las funciones polinómicas tienen la forma: $Pol(x)= a_n x^n+ a_{n-1} x^{n-1}= a_2 x^2+ a_1 x$:

A una función con esta forma se le llama polinomio de grado $n$ (El mayor número).

A los elementos a los llamaremos coeficientes donde: $a ∈R$.

Por ejemplo: $P(x)=2x^7-x^4+3x^2+4$, es un polinomio de grado 7 por el mayor índice en el termino.

```python
## Funcion algebraica polinomica
def f(x):
  return (2*x**7)-(x**4)+(3*x**2)+4
y = f(x)

#>> Grafica de la funcion
plt.plot(x, y, label="$P(x)=2x^7-x^4+3x^2+4$")
plt.title('Funcion polinomica')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(loc='upper left')
plt.grid()

plt.show() #Mostrar grafica
```

#### ***Nota con funciones polinomicas***
Al usar los terminos $P(x)=2x^7-x^4+3x^2+4$ y cambiar el primer exponente a:
> Un numero par, la forma de la funcion dentra una forma de $U$.

> Un numero impar, la forma de la funcion sera una onda ascendente.

### ***Funciones de potencia***
Similares a las funciones polinomicas tienen la forma: $f(x) = x^2$, el

Dominio es $Dom_f = (-∞, ∞)$, y su
imagen es $Im_f = (0, ∞)$

Por ejemplo: $f(x) = x^2$

```python
## Funcion algebraica de potencia
def f(x):
  return x**7
y = f(x)

#>> Grafica de la funcion
plt.plot(x, y, label="$P(x)=x^7$")
plt.title('Funcion de potencia')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(loc='upper left')
plt.grid()

plt.show() #Mostrar grafica
```

### ***Funciones algebraicas trascendentes***
Aquellas funciones que no pueden ser expresadas con polinomios.

#### ***Funciones trigonometricas***
Algunos ejemplos son las funciones: $sin(x), cos(x)$ y $tan(x)$

```python
## Funciones trigonometricas
def f(x):
    return np.sin(x), np.cos(x), np.tan(x)
y_sin, y_cos, y_tan = f(x)

#>> Grafica de la funcion
fig, ax = plt.subplots(3, 1, figsize=(8, 8), label=('Funciones trigonometricas'))

#> Sin(x)
ax[0].plot(x, y_sin, label="$Sin(x)$")
ax[0].set_title('Seno de X')
ax[0].set_xlabel('X')
ax[0].set_ylabel('Y')
ax[0].legend(loc='upper left')
ax[0].grid()

#> Cos(x)
ax[1].plot(x, y_cos, label="$Cos(x)$")
ax[1].set_title('Coseno de X')
ax[1].set_xlabel('X')
ax[1].set_ylabel('Y')
ax[1].legend(loc='upper left')
ax[1].grid()

#> Sin(x)
ax[2].plot(x, y_tan, label="$Tan(x)$")
ax[2].set_title('Tangente de X')
ax[2].set_xlabel('X')
ax[2].set_ylabel('Y')
ax[2].legend(loc='upper left')
ax[2].grid()

fig.tight_layout() #Espaciado entre graficas
plt.show() #Mostrar grafica

```

#### ***Funciones exponenciales***
Tienen la forma de $f(x) = a^x$

Donde la base $a$ es una constante positiva. Un gran ejemplo de una funcion exponencial es usando la base como el numero de $euler$:

$f(x) = e^2$

```python
## Funciones exponenciales
def f(x):
    return np.exp(x)
y = f(x)

#>> Graficas de la funcion
plt.plot(x, y, label='$f(x) = e^2$')
plt.title('Funcion exponencial')
plt.ylabel('Y')
plt.xlabel('X')
plt.legend()
plt.grid()

plt.show()
```

#### ***Funcion logaritmicas***
El logaritmo esta definido por la **relacion**: $log_b(x) = n$ **->** $x = b^n$

Donde:
>$b$ **->** Es la base

>$n$ **->** Es el exponente al que esta elevado la base

>$x$ **->** Es el resultado de elevar la base $b$ al exponente $n$

***Ejemplo:*** Teniendo $b=2$ y $n=8$, entonces:  $2^8 = 256$

Por lo que: $x=256$. Calculando el logaritmo base 2 de $x$ es:

$log_2(256) = 8$

```python
## Funciones logaritmicas
def f(x):
    return np.log2(x)

x = np.linspace(0.001, 256, 1000) #El primer numero debe ser diferente de 0
y = f(x)

#>> Graficas de la funcion
plt.plot(x, y, label='$log_2(256) = 8$')
plt.title('Funcion algoritmica')
plt.ylabel('Y')
plt.xlabel('X')
plt.legend()
plt.grid()

plt.show()
```

### ***Funciones seccionadas***
Son funciones que tienen diferentes valores definidos por un intervalo. Por ejemplo la funcion de escalon de Heaviside:

$H(x) = {{0, x < 0 | 1, x > 0}}$

```python
## Funciones seccionadas
def H(x):
    v = np.zeros(len(x))
    for idx, x in enumerate(x):
        if x > 0:
            v[idx] = 1
        else:
            v[idx] = 0
    return v        
    
n = 1000
x = np.linspace(-10, 10, n)    
        
y = H(x)

#>> Graficas de la funcion
plt.plot(x, y, label='$H(x) = 0, x<0|1, x>0$')
plt.title('Funciones seccionadas o condicionales')
plt.ylabel('Y')
plt.xlabel('X')
plt.legend()
plt.grid()

plt.show()
```

### ***Funciones compuestas***
Las funciones compuestas estan conformadas por dos o mas funciones algebraicas:

$f-o-g = f(g(x))$

```python
## Funciones compuestas
n = 1000
x = np.linspace(-10, 10, n)

def g(x):
    return x**2

def f(x):
    return np.sin(x)

f_o_g = f(g(x)) #Combinar las funciones desde las variables

##> Grafica de las funciones
plt.plot(x, f_o_g, label='f(g(x))')
plt.title('Funciones compuestas')
plt.ylabel('Y')
plt.xlabel('X')
plt.legend()
plt.grid()

plt.show()
```

## ***Manipulacion de funciones***

### ***Desplazamientos horizontales y verticales***


Siendo $c$ una constante mayor que 0, entonces la grafica:

> $y = f(x) + c$, se desplaza $c$ unidades hacia arriba

> $y = f(x) - c$, se desplaza $c$ unidades hacia abajo

> $y = f(x + c)$, se desplaza $c$ unidades hacia derecha

> $y = f(x - c)$, se desplaza $c$ unidades hacia izquierda


```python
## Desplazamientos horizontales y verticales
n = 1000
x = np.linspace(-10, 10, n)
c = 5 #Numero de unidades que nos queremos desplazar

def f(x):
    return x**2
y = f(x) - c

##> Grafica de la funcion
fig, ax = plt.subplots()
ax.plot(x, y,label='$f(x) = m * x + b$')
ax.set_title('Desplazamientos horizontales y verticales')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.grid()
ax.legend()

ax.axvline(x=0, color='g')
ax.axhline(y=0, color='g')

plt.show()
```

### ***Alargamientos y compresiones***

Siendo $c$ una constante mayor que 0, entonces la grafica:

> $y = c * f(x)$, alarga la grafica verticalmente en un factor de $c$.

> $y = \frac{1}{c} * f(x)$, comprime la grafica verticalmente en un factor de $c$.

> $y = f(c * x)$, comprime la grafica horizontalmente en un factor de $c$.

> $y = f(\frac{1}{c} * x)$, alarga la grafica horizontalmente en un factor de $c$.

```python
## Alargamientos y compresiones
n = 1000
def f(x):
    return np.sin(x)

c = 2
x = np.linspace(-15, 15, n)
y = f(1/c * x) #Aqui se hace la operacion para modificar la grafica

#>> Grafica de la funcion
fig, ax = plt.subplots()
ax.plot(x, y, label='$sin(x)$')
ax.set_title("Alargamientos de funciones")
ax.grid()
ax.legend(loc='upper left')
ax.axhline(y=0, color='g')
ax.axvline(x=0, color='g')

plt.show()
```

### ***Reflexiones de funciones***

Reflejar una funcion:

>$y = -f(x)$, refleja la grafica respecto al eje $x$

>$y = f(-x)$, refleja la grafica respecto al eje $y$

```python
## Reflexion de funciones algebraicas
n = 1000
def f(x):
    return x**3

x = np.linspace(-100, 100, n)
y = -f(x) #Aqui se hace la operacion con la funcion para reflejarla

#>> Grafica de las funciones
fig, ax = plt.subplots()
ax.set_title('Reflexion de funciones')
ax.plot(x, y, label='$-f(x^3)$')
ax.grid()
ax.legend()
ax.axhline(y=0, color='g')
ax.axvline(x=0, color='g')

plt.show()
```
