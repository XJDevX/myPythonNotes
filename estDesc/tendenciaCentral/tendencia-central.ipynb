#### Variables del notebook

```python
import pandas as pd
import seaborn as sns

df = pd.read_csv("A. dataSet01.csv")
```

## ***Medidas de tendencia central***
Media: $\frac{1}{N} \sum_{i=1}^{N} x_i$

Mediana (Impar): $x_{(n+1)/2}^{ordered}$

Mediana (Par): $\frac{x_{n/2}^{ordered}+x_{(n+1)/2}^{ordered}}{2}$

Moda: $x_k$ donde $Freq(x_k) = max(Freq(x_1))$

### Media
Es el promedio de los valores del conjunto especificado:

```python
print(f"Media del precio en USD: {df['price_usd'].mean()}")
```

### Mediana
Es el valor del medio del conjunto especificado:

```python
print(f"Mediana del precio en USD: {df['price_usd'].median()}")
df['price_usd'].plot.hist(bins=50)
```

Se puede hacer mas interesante al analizar los precios por marcas:

```python
sns.displot(df, x='price_usd', hue='manufacturer_name', multiple='stack')
```

El diagrama anterior es dificil de analizar, donde estan los autos electricos?

Inspeccionemos precios de una marca y modelo particular:

```python
chevroCaptivaDF = df[(df['manufacturer_name'] == 'Chevrolet') & (df['model_name'] == 'Captiva')]
sns.displot(chevroCaptivaDF, x='price_usd', hue='year_produced')
```
